```python
import pandas as pd

# Load the CSV (upload to Colab first if using Option A)
df = pd.read_csv("retail_store_inventory.csv")

# Show first 5 rows
print("First 5 rows:")
print(df.head())

# Show column names
print("\nColumns in dataset:")
print(df.columns.tolist())

# Shape of dataset
print("\nShape (rows, columns):", df.shape)
```

```python
# Convert Date column to datetime type
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Convert numeric columns
df["Units Sold"] = pd.to_numeric(df["Units Sold"], errors="coerce")
df["Inventory Level"] = pd.to_numeric(df["Inventory Level"], errors="coerce")

# Check if any missing values
print(df.isna().sum())

# Look at basic info
print("\nDataset info:")
print(df.info())
```

```python
# Group by Product and Week
df_weekly = (
    df
    .groupby(["Product ID", pd.Grouper(key="Date", freq="W")], as_index=False)
    .agg({
        "Units Sold": "sum",         # total units sold that week
        "Inventory Level": "last"    # inventory snapshot at the end of the week
    })
    .sort_values(["Product ID", "Date"])
)

print("Weekly dataset shape:", df_weekly.shape)
print("\nFirst 10 rows of weekly data:")
print(df_weekly.head(10))
```

```python
import numpy as np

def forecast_moving_average(series, window=4, horizon=4):
    """
    Simple moving average forecast.
    series: pandas Series of weekly sales
    window: how many past weeks to average
    horizon: how many weeks ahead to forecast
    """
    history = series.dropna().values
    if len(history) < window:
        # Not enough history, just use overall mean
        avg = np.mean(history) if len(history) > 0 else 0
        return [avg] * horizon

    # Rolling average of the last 'window' weeks
    avg = np.mean(history[-window:])
    return [avg] * horizon

# Example: pick one product to test
example_product = df_weekly["Product ID"].unique()[0]
prod_df = df_weekly[df_weekly["Product ID"] == example_product]

forecast = forecast_moving_average(prod_df["Units Sold"], window=4, horizon=4)

print("Product:", example_product)
print("Next 4 weeks forecast:", forecast)
```

```python
# Forecast for every product (using moving average)
results = []

for pid in df_weekly["Product ID"].unique():
    prod_df = df_weekly[df_weekly["Product ID"] == pid].sort_values("Date")
    forecast = forecast_moving_average(prod_df["Units Sold"], window=4, horizon=4)
    forecast_total = sum(forecast)  # total demand over next 4 weeks

    # Get current inventory (last available value)
    current_inv = prod_df["Inventory Level"].iloc[-1]

    results.append({
        "Product ID": pid,
        "Forecast (4 weeks)": round(forecast_total, 2),
        "Current Inventory": round(current_inv, 2)
    })

forecast_summary = pd.DataFrame(results)
print("Forecast summary (first 10 rows):")
print(forecast_summary.head(10))
```

```python
def optimize_inventory(forecast, inventory, safety_factor=1.2, overstock_factor=1.5):
    optimal_stock = forecast * safety_factor
    reorder_qty = max(optimal_stock - inventory, 0)

    if inventory < forecast:
        status = "Stock-out Risk"
    elif inventory > forecast * overstock_factor:
        status = "Overstock Risk"
    else:
        status = "Healthy"

    return round(optimal_stock, 2), round(reorder_qty, 2), status

# Apply optimization to all products
optimized_results = []

for _, row in forecast_summary.iterrows():
    pid = row["Product ID"]
    forecast = row["Forecast (4 weeks)"]
    inventory = row["Current Inventory"]

    optimal_stock, reorder_qty, status = optimize_inventory(forecast, inventory)

    optimized_results.append({
        "Product ID": pid,
        "Forecast (4 weeks)": forecast,
        "Current Inventory": inventory,
        "Optimal Stock": optimal_stock,
        "Reorder Quantity": reorder_qty,
        "Status": status
    })

decision_table = pd.DataFrame(optimized_results)
print("Inventory Decision Table (first 10 rows):")
print(decision_table.head(10))
```

```python
import matplotlib.pyplot as plt

def plot_product(product_id):
    # Get weekly data for this product
    prod_df = df_weekly[df_weekly["Product ID"] == product_id].sort_values("Date")

    # Sales history
    sales = prod_df["Units Sold"].values
    dates = prod_df["Date"].values

    # Inventory history
    inv = prod_df["Inventory Level"].values

    # Forecast from decision table
    row = decision_table[decision_table["Product ID"] == product_id].iloc[0]
    forecast_total = row["Forecast (4 weeks)"]
    weekly_forecast = forecast_total / 4  # evenly spread
    future_dates = pd.date_range(prod_df["Date"].max(), periods=4, freq="W")
    forecast_values = [weekly_forecast] * 4

    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(dates, sales, label="Weekly Sales", marker="o")
    plt.plot(dates, inv, label="Inventory Level", marker="s")
    plt.plot(future_dates, forecast_values, label="Forecast (next 4 weeks)", linestyle="--")

    plt.title(f"Product {product_id} â€” Sales, Inventory & Forecast")
    plt.xlabel("Date")
    plt.ylabel("Units")
    plt.legend()
    plt.show()

# Example: plot first product
plot_product(decision_table.iloc[0]["Product ID"])
```

```python
# Save the decision table to CSV
decision_table.to_csv("inventory_decision_table.csv", index=False)
print("Saved: inventory_decision_table.csv")
```
